## Data Cleaning & Preprocessing

```python
# Create working copy
df_clean = df.copy()

# --- Parse timestamp ---
if 'timestamp' in df_clean.columns:
    ts_col = 'timestamp'
elif 'Timestamp' in df_clean.columns:
    ts_col = 'Timestamp'
elif 'Date_Time' in df_clean.columns:
    ts_col = 'Date_Time'
else:
    raise KeyError("‚ùå No timestamp column found! Please check your dataset.")

df_clean[ts_col] = pd.to_datetime(df_clean[ts_col], errors='coerce')
df_clean = df_clean.sort_values(ts_col).reset_index(drop=True)

# --- Identify consumption column ---
if 'Electricity_Consumed' in df_clean.columns:
    consumption_col = 'Electricity_Consumed'
elif 'consumption' in df_clean.columns:
    consumption_col = 'consumption'
elif 'Global_active_power' in df_clean.columns:
    consumption_col = 'Global_active_power'
else:
    raise KeyError(f"‚ùå No valid consumption column found! Available columns: {list(df_clean.columns)}")

# --- Handle missing values in consumption column ---
# Forward fill for small gaps, then interpolate
df_clean[consumption_col] = df_clean[consumption_col].fillna(method='ffill').fillna(method='bfill')
df_clean[consumption_col] = df_clean[consumption_col].interpolate(method='linear')

# Remove remaining nulls
df_clean = df_clean.dropna(subset=[consumption_col])

# --- Remove outliers using IQR method ---
Q1 = df_clean[consumption_col].quantile(0.25)
Q3 = df_clean[consumption_col].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_before = len(df_clean)
df_clean = df_clean[(df_clean[consumption_col] >= lower_bound) & 
                    (df_clean[consumption_col] <= upper_bound)]
outliers_removed = outliers_before - len(df_clean)

print(f"‚úÖ Outliers removed: {outliers_removed}")
print(f"‚úÖ Data shape after cleaning: {df_clean.shape}")
print(f"‚úÖ Missing values remaining: {df_clean[consumption_col].isnull().sum()}")
```

## Create Temporal Features

```python
# --- Get timestamp column name ---
if 'Timestamp' in df_clean.columns:
    ts_col = 'Timestamp'
elif 'timestamp' in df_clean.columns:
    ts_col = 'timestamp'
elif 'Date_Time' in df_clean.columns:
    ts_col = 'Date_Time'
else:
    raise KeyError("‚ùå No timestamp column found in the dataset.")

# --- Extract time features ---
df_clean['year'] = df_clean[ts_col].dt.year
df_clean['month'] = df_clean[ts_col].dt.month
df_clean['day'] = df_clean[ts_col].dt.day
df_clean['hour'] = df_clean[ts_col].dt.hour
df_clean['dayofweek'] = df_clean[ts_col].dt.dayofweek  # 0=Monday, 6=Sunday
df_clean['week'] = df_clean[ts_col].dt.isocalendar().week.astype(int)
df_clean['quarter'] = df_clean[ts_col].dt.quarter
df_clean['dayofyear'] = df_clean[ts_col].dt.dayofyear

# --- Categorical flags ---
df_clean['is_weekend'] = (df_clean['dayofweek'] >= 5).astype(int)
df_clean['is_holiday'] = 0  # Can be enhanced later with actual holiday dates

# --- Cyclic encoding for hour and day of week ---
df_clean['hour_sin'] = np.sin(2 * np.pi * df_clean['hour'] / 24)
df_clean['hour_cos'] = np.cos(2 * np.pi * df_clean['hour'] / 24)
df_clean['day_sin'] = np.sin(2 * np.pi * df_clean['dayofweek'] / 7)
df_clean['day_cos'] = np.cos(2 * np.pi * df_clean['dayofweek'] / 7)

print("‚úÖ Temporal features created:")
print(df_clean[['year', 'month', 'day', 'hour', 'dayofweek', 'is_weekend', 'hour_sin', 'hour_cos']].head())
```

## Create Lagged & Rolling Features

```python
# Identify the correct consumption column
if 'Electricity_Consumed' in df_clean.columns:
    consumption_col = 'Electricity_Consumed'
elif 'consumption' in df_clean.columns:
    consumption_col = 'consumption'
elif 'Global_active_power' in df_clean.columns:
    consumption_col = 'Global_active_power'
else:
    raise KeyError(f"No valid consumption column found! Available columns: {list(df_clean.columns)}")

# --- Lagged features ---
lags = [1, 6, 12, 24, 48, 168]  # 1 hour, 6 hours, 12 hours, 24 hours, 2 days, 1 week

for lag in lags:
    df_clean[f'{consumption_col}_lag_{lag}'] = df_clean[consumption_col].shift(lag)

# --- Rolling statistics (mean, std, min, max) ---
windows = [6, 12, 24, 168]  # 6 hours, 12 hours, 24 hours, 1 week

for window in windows:
    df_clean[f'{consumption_col}_rolling_mean_{window}'] = df_clean[consumption_col].rolling(window).mean()
    df_clean[f'{consumption_col}_rolling_std_{window}'] = df_clean[consumption_col].rolling(window).std()
    df_clean[f'{consumption_col}_rolling_min_{window}'] = df_clean[consumption_col].rolling(window).min()
    df_clean[f'{consumption_col}_rolling_max_{window}'] = df_clean[consumption_col].rolling(window).max()

# --- Drop rows with NaN from lag/rolling features ---
df_clean = df_clean.dropna().reset_index(drop=True)

print("‚úÖ Feature engineering complete!")
print(f"Total features created: {len(df_clean.columns)}")
print(f"Final dataset shape: {df_clean.shape}")
print("\nFeature columns:")
print(df_clean.columns.tolist())
```

## Deep Learning Model - LSTM

```python
# Import TensorFlow/Keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

print("Preparing data for LSTM...")

# Create sequences for LSTM
def create_sequences(data, seq_length=24):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length])
    return np.array(X), np.array(y)

# Use scaled consumption data
y_scaled = (y - y.min()) / (y.max() - y.min())

seq_length = 24
X_lstm, y_lstm = create_sequences(y_scaled.values, seq_length)

# Split (80-20)
split_idx_lstm = int(len(X_lstm) * 0.8)
X_train_lstm = X_lstm[:split_idx_lstm]
y_train_lstm = y_lstm[:split_idx_lstm]
X_test_lstm = X_lstm[split_idx_lstm:]
y_test_lstm = y_lstm[split_idx_lstm:]

print(f"LSTM Training shape: {X_train_lstm.shape}")
print(f"LSTM Test shape: {X_test_lstm.shape}")

print("\nTraining LSTM Model...")
print("="*60)

try:
    # Build LSTM model
    model_lstm = Sequential([
        LSTM(64, activation='relu', input_shape=(seq_length, 1), return_sequences=True),
        Dropout(0.2),
        LSTM(32, activation='relu'),
        Dropout(0.2),
        Dense(16, activation='relu'),
        Dense(1, activation='sigmoid')
    ])
    
    model_lstm.compile(optimizer=Adam(learning_rate=0.001), loss='mse', metrics=['mae'])
    
    # Early stopping
    early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
    
    # Train (suppress output)
    history = model_lstm.fit(
        X_train_lstm, y_train_lstm,
        epochs=20,
        batch_size=32,
        validation_split=0.1,
        callbacks=[early_stop],
        verbose=0
    )
    
    # Predictions
    y_pred_lstm_scaled = model_lstm.predict(X_test_lstm, verbose=0)
    y_pred_lstm = y_pred_lstm_scaled * (y.max() - y.min()) + y.min()  # Unscale
    
    results['LSTM'] = calculate_metrics(y_test_lstm * (y.max() - y.min()) + y.min(), y_pred_lstm)
    print(f"LSTM: {results['LSTM']}")
    
except Exception as e:
    print(f"LSTM Error: {e}")
    results['LSTM'] = {'MAE': np.nan, 'RMSE': np.nan, 'MAPE': np.nan, 'R2': np.nan}

print("\n‚úÖ LSTM model complete")
```
